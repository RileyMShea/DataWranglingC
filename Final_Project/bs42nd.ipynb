```python
# -*- coding: utf-8 -*-
import cchardet as chardet
with open(r"rochester_ny.osm", "rb") as f:
    msg = f.read()
    result = chardet.detect(msg)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(msg, 'xml')
```

```python
nodes = soup.find_all('node')

ways = soup.find_all('way')
```

```python
node_attr_keys = dict()
for entry in nodes:
    if len(entry.contents) > 0:
        for k in entry.attrs.keys():
            if k not in node_attr_keys.keys():
                node_attr_keys[k] = 1
            else:
                node_attr_keys[k] += 1

pprint("keys: counts")
pprint(node_attr_keys)
```

```python
way_attr_keys = dict()
for entry in ways:
    if len(entry.contents) > 0:
        for k in entry.attrs.keys():
            if k not in way_attr_keys.keys():
                way_attr_keys[k] = 1
            else:
                way_attr_keys[k] += 1

pprint("keys: counts")
pprint(way_attr_keys)
```
